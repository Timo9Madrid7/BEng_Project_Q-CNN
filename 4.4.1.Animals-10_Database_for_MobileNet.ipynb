```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import matplotlib.pyplot as plt
import numpy as np
```

```python
# example of data
train_data_path = './Animals_datasets/raw-img'

img = load_img(train_data_path+'/elefante/e032b10b2df21c22d2524518b7444f92e37fe5d404b0144390f8c07aa5ecb6_640.jpg',
               target_size=(224,224))
plt.imshow(img)
plt.show()
img_array = img_to_array(img)
img_array_exp = np.expand_dims(img_array, axis=0)
print()
print("the shape of the above image:", img_array_exp.shape)
```

```python
# data augmentation
# split the data to two categories, one for testing,and another one for validating
ImageDataGenerator()

train_data = ImageDataGenerator(rescale=1.0/255.,
                                rotation_range=40,
                                width_shift_range=0.2,height_shift_range=0.2,
                                validation_split=0.2,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True,
                               ) 

train_generator = train_data.flow_from_directory(train_data_path,
                                                target_size=(224,224),
                                                batch_size=8,
                                                class_mode = 'categorical',
                                                subset='training'
                                                )

validation_generator = train_data.flow_from_directory(train_data_path,
                                                target_size=(224,224),
                                                batch_size=8,
                                                class_mode = 'categorical',
                                                subset='validation'
                                                )
```
